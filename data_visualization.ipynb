# NumPy

### 1. 資料生成(Data Generation)

- `np.zeros(shape)`:生成元素全為$0$的陣列，引數(Argument)為形狀(Shape)數組(Tuple)。
- `np.ones(shape)`生成元素全為$1$的陣列，引數(Argument)為形狀(Shape)數組(Tuple)。
- 利用逐元素乘法(Element-wise multiplication)搭配`np.ones(shape)`可以創建任何元素相同的陣列。
- `np.eye(number)`或`np.identity(number)`:創建單位(Identity)矩陣，引數(Argument)為列數。
- `np.arange(start, stop, step)`用法和Python的`range(start, stop, step)`一樣，可以生成等差數列。
- `np.linspace(start, stop, num=50)`是另一種生成等差數列的方法，指定總共元素的數量，預設值為$50$。

### 2. 亂數產生器(Random Generator)

- `RandomState`: 舊的亂數生成(Legacy Random Generation)
  - `np.random.random(shape)`或`np.random.rand()`:根據連續均勻分布(Continous Uniform Distribution)生成亂數，可以將形狀(Shape)當成引數(Argument)來指定陣列形狀。注意後者參數不是數組(Tuple)，類似`reshape()`。
  - `np.random.randn()`:根據標準高斯分布(Normal Distribution)生成亂數。
  - `np.random.normal(loc=0.0, scale=1.0, size=shape)`:根據高斯分布(Standard Normal Distribution)生成亂數，`loc`是平均(Mean)，`scale`是標準差(Standard Deviation)。
  - `np.random.randint(low, high, shape)`:根據離散均勻分布(Discrete Uniform Distribution)生成亂數，`low`指定下限，`high`指定上限，關鍵字引數(Keyword Argument)`size`則是形狀(Shape)。
  - `np.random.choice(int, shape)`:根據離散均勻分布(Discrete Uniform Distribution)生成亂數，第一個參數是`range(int)`的意思。
  - `np.random.seed(seed)`和`np.random.RandomState(seed)`:可以指定亂數種子(Seed)，確保每次生成結果一樣，使用後者可以避免全域狀態(Global State)。
- `Generator`:新的亂數產生器(Random Generator)
  - 函數和舊的大致相同，沒有`rand()`和`randn()`這兩個快速用法。
  - `np.random.integers(low, high, shape)`取代舊的`np.random.randint(low, high, shape)`。
  - `np.random.default_rng(seed)`:可以指定亂數種子(Seed)，確保每次生成結果一樣。

### 3. 探索式資料分析(Exploratory Data Analysis, EDA)
是一套包含資料視覺化及統計等知識和技術的數據分析方法；其主要目的是從各個面向探索數據，找出解決問題的線索並進一步確認問題及解決方法。探索式資料分析通常沒有可依循的步驟，而是以一連串的啟發式的提問-回答的循環、逐漸的釐清問題的核心，從數據中挖掘對問題深度的見解。

- 方法(Methods)
  - `df.head()`

  - `df.info()`

### 4. 資料的連接(Join)與合併(Merge)
- `df.concat(ignore_index, axis)`：連接兩個DataFrame，可以指定`axis`，$0$從下方加上列(Row)，$1$從右邊加上行(Column)。`ignore_index=True`可以重編索引標籤(Index Label)。
  - `df.append()`：舊的函數，只支援從下方加上列。
- `df.merge(how="inner, on, left_on, right_on)`：資料庫風格合併(Database-Style Join)。`how`支援`left`、`right`、`outer`、`inner`、`cross`五種合併方式。`on`行名(Column Name)作為合併用的鍵(Key)，也可以用`left_on`和`right_on`分別指定。
- `df.join(how="left", on)`：利用索引標籤(Index Label)來合併。`how`也同樣支援各種合併(Join)。`on`輸入原DataFrame的列名(Column Name)作為鍵(Key)，用來對對象DataFrame的索引標籤(Index Label)執行合併(Join)。

```python
left = pd.DataFrame(
  {
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"],
    "C": ["C0", "C1", "C2"]
  }
)
left
```

```python
right = pd.DataFrame(
  {
    "A": ["A3", "A4", "A5"],
    "B": ["B3", "B4", "B5"],
    "C": ["C3", "C4", "C5"]
  }
)
right
```

```python
# `concat()`
pd.concat([left, right])
```

```python
# Ignore index
pd.concat([left, right], ignore_index=True)
```

```python
# Set axis to concatenate
pd.concat([left, right], axis=1)
```

```python
# `merge()`
left = pd.DataFrame(
  {
    "Key": ["K0", "K1", "K2"],
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"],
  }
)

right = pd.DataFrame(
  {
    "Key": ["K0", "K1", "K2"],
    "C": ["C0", "C1", "C2"],
    "D": ["D0", "D1", "D2"],
  }
)

pd.merge(left, right, on="Key")
```

```python
left = pd.DataFrame(
  {
    "Key1": ["K0", "K1", "K2"],
    "Key2": ["K1", "K2", "K3"],
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"],
  }
)
left
```

```python
right = pd.DataFrame(
  {
    "Key1": ["K1", "K2", "K3"],
    "Key2": ["K2", "K3", "K4"],
    "C": ["C0", "C1", "C2"],
    "D": ["D0", "D1", "D2"],
  }
)
right
```

```python
# Inner join using column name
pd.merge(left, right, on="Key1")
```

```python
pd.merge(left, right, on="Key2")
```

```python
pd.merge(left, right, left_on="Key1", right_on="Key2")
```

```python
# Inner join using list
pd.merge(left, right, on=["Key1", "Key2"])
```

```python
# Left join
pd.merge(left, right, how="left", on=["Key1", "Key2"])
```

```python
# Right join
pd.merge(left, right, how="right", on=["Key1", "Key2"])
```

```python
# Outer join
pd.merge(left, right, how="outer", on=["Key1", "Key2"])
```

```python
# Cross join
pd.merge(left, right, how="cross")
```

```python
left = pd.DataFrame(
  {
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"]
  },
  index=["K0", "K1", "K2"]
)
left
```

```python
right = pd.DataFrame(
  {
    "C": ["C0", "C1", "C2"],
    "D": ["D0", "D1", "D2"]
  },
  index=["K1", "K2", "K3"]
)
right
```

```python
# `join(how="left")`
left.join(right)
```

```python
# Outer join
left.join(right, how="outer")
```

```python
left.join(right, on="A")
```

```python
left.join(right, how="outer", on="A")
```

### 8. 設定座標軸標籤(Labels)、標題(Title)

- `plt.xlabel()`和`plt.ylabel()`設定座標軸標籤。
- `plt.title()`設定標題。

```python
# `xlabel`, `ylabel`, and `title`
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

X = np.linspace(-20, 20, 100)
Y = sigmoid(X)

plt.plot(X, Y)
plt.xlabel("Input")
plt.ylabel("Uutput")
plt.title("Sigmoid Function");
```

```python
# Custom fonts
def relu(x):
    return(np.maximum(0, x))

x = np.linspace(-20, 20, 100)
y = relu(X)

custom_font = {"family": "serif", "color": "#008080", "size": 10}

plt.xlabel("Input")
plt.ylabel("Output")
plt.title('ReLU', fontdict=custom_font)
plt.plot(x, y);
```

```python
# Markers
x = np.arange(-10, 10)
y = np.tanh(x)

plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Hyperbolic Tangent Function")
plt.plot(x, y, marker = "o");
```

### 10. 繪製多個子圖(Subplots)

`plt.subplot(nrows, ncols, index)`繪製多個子圖，`nrows`代表列(Row)數，`ncols`代表行(Column)數，`index`則是這張圖是第幾張。

```python
# `subplot()`
x = np.linspace(-1, 1, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.arcsin(x)
y5 = np.arccos(x)
y6 = np.arctan(x)

plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.title("sin(x)")

plt.subplot(2, 3, 2)
plt.plot(x, y2)
plt.title("cos(x)")

plt.subplot(2, 3, 3)
plt.plot(x, y3)
plt.title("tan(x)")

plt.subplot(2, 3, 4)
plt.plot(x, y4)
plt.title("arcsin(x)")

plt.subplot(2, 3, 5)
plt.plot(x, y5)
plt.title("arccos(x)")

plt.subplot(2, 3, 6)
plt.plot(x, y6)
plt.title("arctan(x)")

plt.suptitle("Trigonometric Functions")
plt.tight_layout();
```
