# 1. Financial Statements
Only annual & quarterly historical balance sheet data are available on Yahoo.
## 1-1. Balance Sheet

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
aapl = yf.Ticker(ticker="AAPL")
aapl
```

```python
# Annual in million USD
balance_sheet = aapl.get_balance_sheet() / 1000000
balance_sheet
```

## 1-2. Income Statement

```python
# In million USD
income_stmt = aapl.get_income_stmt() / 1000000
income_stmt
```

## 1-3. Cashflow Statement

```python
# In million USD 
cashflow = aapl.get_cashflow() / 1000000
cashflow
```
